```python
marks = int(input("Enter the student's score: "))

if marks <= 100 and marks >= 90:
    print("Grade A")
elif marks < 90 and marks >= 80:
    print("Grade B")
elif marks < 80 and marks >= 70:
    print("Grade C")
elif marks < 70 and marks >= 50:
    print("Grade D")
elif marks < 50 and marks >= 0:
    print("Fail")
else:
    print("Invalid score")
```

```python
signal = input("Enter traffic light color (red / yellow / green): ")
if signal == "red":
    print("STOP")
elif signal == "yellow":
    print("GET READY")
elif signal == "green":
    print("GO")
else:
    print("Invalid traffic signal")
```

```python
users = {
    "sara": "pass123",
    "samee": "abc456",
    "rahamath": "abc789",
    "mohana": "python123",
    "kiran": "toptech3"
}

user_id = input("Enter User ID: ")
password = input("Enter Password: ")

if user_id in users:
    if users[user_id] == password:
        print("Login successful")
    else:
        print("Incorrect password ")
else:
    print("User ID not found")
    choice = input("Do you want to create a new account? (yes/no): ")

    if choice.lower() == "yes":
        new_user = input("Create new User ID: ")
        new_pass = input("Create new Password: ")
        users[new_user] = new_pass
        print("New account created successfully ")
    else:
        print("Thank you!")
```
